Operators

```python
n1 = float(input("Enter First Number: "))
n2 = float(input("Enter Second Number: "))

print(f"The SUM is: {n1+n2}")
print(f"The Difference is: {n1-n2}")
print(f"The Product is: {n1*n2}")
print(f"The Division is: {n1/n2}")
print(f"The Modulus is: {n1%n2}")
print(f"The Power of {n1} to {n2} is {n1**n2}")

if n1>n2:
    print("N1 Greater")
elif n2>n1:
    print("N2 Greater")
else:
    print("Egual or Invalid")
```

Conditionals

```python
age = float(input("Enter Your Age Dear: "))

def types(age):
    if age < 18:
        return "Minor"
    elif age >= 18 and age <= 59:
        return "Adult"
    elif age > 59:
        return "Senior Citizen"

print(f"So, You're age is {age} Right? You look", types(age))
```

Login System

```python
username = "anas"
password = "1234"
count = 3

while count > 0:

    user = input("Enter Username: ")
    pas = input("Enter Password: ")

    if user == username and pas == password:    
        print("Login Successful")
        break
    else:
        print("Invalid Credentials")
        count-=1
    
```
