Linear Search

```python
rolls = [ 56, 57, 58, 59, 60, 61, 62, 63, 64, 67]

def linear_search(arr, k):

    for roll in arr:
        if roll == k:
            return "Congrats! You're on the list"

    return "Sorry! You're not in the list"        
        
for roll in rolls: 
    print(roll)

print("The Length of List is: ", len(rolls))

key = int(input("Whats Your Roll Number: "))

print(linear_search(rolls, key))


```

Marks

```python
marks = [78, 90, 65, 88, 95]
total = 0

print("Max marks in class are: ", max(marks))
print("Min marks in class are: ", min(marks))

for i in marks:
    total += i+1

print("The Avg of class is: ", total/len(marks))

print(marks)
marks.sort()
print(marks)
marks.reverse()
print(marks)
```

Mini Project 1: Student Manager

```python
import sys

students = []

def add_student():
    addname = input("Enter his name: ")
    students.append(addname)
    print("Student added Succesfully!")
    print("\nCurrent List: ")
    print(students)

def rm_student():
    
    rmname = input("Enter his name: ")
    for names in students:
        if rmname in students:
            students.remove(rmname)
            print("Student Removed Succesfully!")
            print("\nCurrent List:")
            print(students)
            break
        else:
            print("Name isn't in the list, Try Again!")
            rm_student()

while True:

    print(" ==== Welcome to Student Manager Portal ==== ")
    print(" == 1. Add Student ")
    print(" == 2. Remove Student ")
    print(" == 3. Show Students ")
    print(" == 4. Exit ")

    choice = int(input("Enter Your Choice Dear: "))

    if choice == 1:
        add_student()

    elif choice == 2:
        rm_student()
    
    elif choice == 3:
        print(students)

    elif choice == 4:
        print("Good Bye, CYA")
        sys.exit(0)

    else:
        print("Invalid Input Bruh! Try Again")
```
