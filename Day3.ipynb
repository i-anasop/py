```python
n = int(input("From upto what you want sum? "))
total = 0

for i in range(n):
    total +=  i+1

print(total)
```

Linear Search

```python
rolls = [ 56, 57, 58, 59, 60, 61, 62, 63, 64, 67]

def linear_search(arr, k):

    for roll in arr:
        if roll == k:
            return "Congrats! You're on the list"

    return "Sorry! You're not in the list"        
        
for roll in rolls: 
    print(roll)

print("The Length of List is: ", len(rolls))

key = int(input("Whats Your Roll Number: "))

print(linear_search(rolls, key))


```
